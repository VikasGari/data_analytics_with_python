importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Loading
Loading the dataset using pandas and Displaying the first few rows using .head().

```python
df = pd.read_csv("student-mat.csv", sep=";")
print(df.head())
```

check for missing values

```python
print("Missing Values:\n", df.isnull().sum())
print("\nColumn Data Types:\n", df.dtypes)
print("\nDataset Shape:", df.shape)
```

c. Data Cleaning
● Handle missing values (e.g., replace them with the median or remove rows).
● Remove duplicate entries using .drop_duplicates().

```python
# we have no missing value in any numerical column
num_columns = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3']
median_values = {col: df[col].median() for col in num_columns}
df.fillna(value=median_values, inplace=True)
```

dropping rows with missing values

```python
df = df.dropna()
```

removing duplicate entries and checking new shape

```python
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

# Data Analysis Questions

1. What is the average score in math (G3)?

```python
avg_g3 = df['G3'].mean()
print("Average math score (G3):", avg_g3)
```

2. How many students scored above 15 in their final grade (G3)?

```python
students_above_15 = df[df['G3'] > 15].shape[0]
print("Number of students with G3 > 15:", students_above_15)
```

3. Is there a correlation between study time (study time) and the final grade (G3)?

```python
correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and G3:", correlation)
```

4. Which gender has a higher average final grade (G3)?

```python
average_by_gender = df.groupby('sex')['G3'].mean()
print("Average G3 by gender:\n", average_by_gender)
```

# Data Visualization
1. Plot a histogram of final grades (G3).
2. Create a scatter plot between study time (study time) and final grade (G3).
3. Create a bar chart comparing the average scores of male and female students.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='studytime', y='G3', hue='sex')
plt.title('Study Time vs Final Grade (G3)')
plt.xlabel('Study Time (hours/week)')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='sex', y='G3', errorbar=None, hue='sex', palette='pastel', legend=False)
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.show()
```
