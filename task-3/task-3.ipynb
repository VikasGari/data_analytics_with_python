# Task 3: Customer Segmentation using Clustering

## Step 1: Load and Inspect the Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

### Load dataset

```python
df = pd.read_csv("customer_data.csv")
```

### Inspect Dataset

```python
print("Dataset shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
print("\nData types:\n", df.dtypes)
df.describe()
df.head()
```

### Data Cleaning Notes

- The dataset has **no missing values** in the columns used for clustering: `Age`, `Annual Income`, and `Spending Score`.
- `Profession` has 35 missing entries but is not used for clustering, so it was ignored.
- **No duplicate rows** were found.

Therefore, **no data cleaning was needed** for this task.


## Step 2: Data Preprocessing

### Selecting relevant features

```python
features = ['Age', 'Annual Income', 'Spending Score']
X = df[features]
```

### Scale features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Step 3: Determine Optimal Clusters for K-Means

### Elbow Method (WCSS)

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

❌ No Clear Elbow Point

### Silhouette Score for evaluation

```python
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    preds = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, preds)
    print(f"Silhouette Score for k={k}: {score:.3f}")
```

✅ Best Silhouette Score: 0.283 at k = 6 and k = 7

### Apply K-Means Clustering (k=6)

```python
optimal_k = 6
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

df.head()
```

## Step 4: Visualize Clusters

### Reduce to 2D for visualization using PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

df.head()
```

### Scatter plot of clusters

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2', s=100)
plt.title('Customer Segments (PCA Reduced)')
plt.grid(True)
plt.show()
```

### t-SNE Visualization of Clusters

While PCA provides a linear projection of the data, t-SNE offers a nonlinear dimensionality reduction technique that can better reveal complex cluster structures.


```python
X_tsne = TSNE(n_components=2, random_state=42, perplexity=30).fit_transform(X_scaled)

df['TSNE1'] = X_tsne[:, 0]
df['TSNE2'] = X_tsne[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='TSNE1', y='TSNE2', hue='Cluster', palette='Set1', s=100)
plt.title('t-SNE Visualization of Customer Clusters')
plt.grid(True)
plt.show()
```

### Pair Plot by Cluster

```python
sns.pairplot(df[features + ['Cluster']], hue='Cluster', palette='husl')
plt.suptitle("Pair Plot of Features by Cluster", y=1.02)
plt.show()
```

### Cluster Profiling

```python
cluster_profile = df.groupby('Cluster')[['Age', 'Annual Income', 'Spending Score']].mean().round(2)
cluster_profile['Customer Count'] = df['Cluster'].value_counts().sort_index()

print("\nMean stats for each cluster:")
cluster_profile
```
